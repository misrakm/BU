```python
import numpy as np
import seaborn as sns
import pandas as pd
```

```python
numdates = 12
np.random.seed(0)
time_series = 100 * (np.random.normal(size = numdates) / 5 + np.arange(numdates) / 16)
datearray = pd.date_range(start='2023/01/01', end='2024/01/01', freq='M')
ax = sns.lineplot(x = datearray, y = time_series)
ax.set_xlabel("Year and month")
ax.set_ylabel("Nitrate level (% of eutrophic value)")
```

Using color intensity change

```python
# Took Help from Google Gemini and made own modifications

import matplotlib.pyplot as plt

def highlight_last_months_intensity(numdates=12, last_months=3):
    np.random.seed(0)
    time_series = 100 * (np.random.normal(size=numdates) / 5 + np.arange(numdates) / 16)
    datearray = pd.date_range(start='2023/01/01', end='2024/01/01', freq='M')
    df = pd.DataFrame({'date': datearray, 'value': time_series})

    # Create the plot
    plt.figure(figsize=(10, 6))


    plt.plot(df['date'], df['value'], color='skyblue')

    # Plot the highlighted last months with increased intensity
    plt.plot(df['date'][-last_months:], df['value'][-last_months:], color='darkblue', linewidth=2.5) #darker blue, thicker line.

    plt.xlabel("Year and month")
    plt.ylabel("Nitrate level (% of eutrophic value)")
    plt.title("Time Series with Highlighted Last Months (Color Intensity)")
    plt.grid(True)
    plt.show()

# Example usage:
highlight_last_months_intensity()
```

Using Color Hue Change

```python
import seaborn as sns

def highlight_last_months_hue(numdates=12, last_months=3):
    np.random.seed(0)
    time_series = 100 * (np.random.normal(size=numdates) / 5 + np.arange(numdates) / 16)
    datearray = pd.date_range(start='2023/01/01', end='2024/01/01', freq='M')
    df = pd.DataFrame({'date': datearray, 'value': time_series})

    # Create the plot
    plt.figure(figsize=(10, 6))

    # Plot the initial data (using blue)
    plt.plot(df['date'], df['value'], color='blue')

    # Plot the highlighted last months with a different hue (using red)
    plt.plot(df['date'][-last_months:], df['value'][-last_months:], color='red', linewidth=2) #red and slightly thicker

    plt.xlabel("Year and month")
    plt.ylabel("Nitrate level (% of eutrophic value)")
    plt.title("Time Series with Highlighted Last Months (Color Hue)")
    plt.grid(True)
    plt.show()

# Example usage:
highlight_last_months_hue()
```

Adding extra markers

```python
def highlight_last_months_markers(numdates=12, last_months=3):
    np.random.seed(0)
    time_series = 100 * (np.random.normal(size=numdates) / 5 + np.arange(numdates) / 16)
    datearray = pd.date_range(start='2023/01/01', end='2024/01/01', freq='M')
    df = pd.DataFrame({'date': datearray, 'value': time_series})

    # Create the plot
    plt.figure(figsize=(10, 6))

    # Plot the initial data (line only)
    plt.plot(df['date'], df['value'], color='skyblue')

    # Plot the highlighted last months with markers
    plt.plot(df['date'][-last_months:], df['value'][-last_months:], color='skyblue', marker='o', linestyle='-') #added markers

    plt.xlabel("Year and month")
    plt.ylabel("Nitrate level (% of eutrophic value)")
    plt.title("Time Series with Highlighted Last Months (Markers)")
    plt.grid(True)
    plt.show()

# Example usage:
highlight_last_months_markers()
```
